```python
## Loading and cleaning Banking_customer_data using pandas

import pandas as pd

df = pd.read_csv('Banking_customer_data.csv')
```

```python
## Converting all column to lowercase 

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'name':'account_holder'})

## Checking the column names 

df.columns
```

```python
## create a new column, age_group

labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
## Importing numpy library for datetime calculation

import numpy as np
from datetime import datetime

##   Convert 'Joined Bank' to datetime objects
##  The format %d/%m/%Y matches '06/05/2019'

df['joined_bank'] = pd.to_datetime(df['joined_bank'], format='%d/%m/%Y')

##  Calculate the tenure in years
##  We calculate the number of days since joining and divide by 365.25

today = pd.Timestamp.now()
df['years_since_joined'] = ((today - df['joined_bank']).dt.days / 365.25).astype(int)

##  Define the grouping logic
##  Grouping bank customers into older, intermediate and new customers

conditions = [
    (df['years_since_joined'] > 10),                                
    (df['years_since_joined'] >= 3) & (df['years_since_joined'] <= 10), 
    (df['years_since_joined'] < 3)                                 
]

choices = ['older_customers', 'intermediate_customer', 'new_customers']

##  Create the new column

df['customer_segment'] = np.select(conditions, choices, default='Unknown')

##  Display the first few rows to verify

print(df[['account_holder', 'joined_bank', 'years_since_joined', 'customer_segment']].head())
```
